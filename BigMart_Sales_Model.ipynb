```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/Gen AI/BigMart Sales Prediction/train_preprocessed.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data.drop(['Item_Identifier', 'Outlet_Identifier'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
X = data.drop(['Item_Outlet_Sales'], axis=1)
y = data['Item_Outlet_Sales']
```

```python
X.shape, y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=10,test_size=0.2)
```

```python
(X_train.shape, y_train.shape), (X_test.shape, y_test.shape)
```

```python
import keras
print(keras.__version__)
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from keras.models import Sequential
from keras.layers import InputLayer, Dense
```

```python
input_neurons = X_train.shape[1]
```

```python
output_neurons = 1
```

```python
number_of_hidden_layers = 2
neuron_hidden_layer_1 = 8
neuron_hidden_layer_2 = 5
```

```python
# defining the architecture of the model
model = Sequential()
model.add(InputLayer(input_shape=(input_neurons,)))
model.add(Dense(units=neuron_hidden_layer_1, activation='relu'))
model.add(Dense(units=neuron_hidden_layer_2, activation='relu'))
model.add(Dense(units=output_neurons))
```

```python
model.summary()
```

```python
model.compile(loss='mean_absolute_error', optimizer='Adam', metrics=['mean_absolute_error'])
```

```python
model_history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50)
```
